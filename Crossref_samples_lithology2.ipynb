```python
arcpy.env.workspace = 'Database Connections\\KL_Exploration.sde'
domain = arcpy.da.ListDomains()[14]
```

```python
data['Geol']= data['Code'].map(domain.codedValues)
```

```python
data = data[(data.SG != '>20.0')]
```

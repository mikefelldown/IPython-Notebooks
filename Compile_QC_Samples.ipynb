```python
 def process(xls):
     
    wb = openpyxl.load_workbook(xls, data_only = True)
    ws = wb.get_sheet_by_name("Descriptions")
    holeid = re.search('\w+\d+[-_]\d+[WEA]?\d?',ws['A1'].value).group(0).replace('-','_')
    test = re.compile('((?<=_)\d+)')
    if int(re.search(test,holeid).group(0)) < 100:
        holeid = re.sub(test,'0'+ re.search(test,holeid).group(1),holeid)
    max_row = ws.max_row
    last_from = 0
    last_to = 0

    for r in ws['H3':'M'+ str(max_row)]:
        #print r[0].value, type(r[1].value)
        if type(r[0]) == float:
            sampleID = "%d" % (r[0].value)
        else:
            sampleID = str(r[0].value)
        if r[1].value and type(r[1].value) == float:
            last_from = r[1].value
            last_to = r[2].value
        if r[5].value and (r[5].value) <> '-':
            val = str(r[5].value)
        #elif r[0].value and type(r[1].value) == unicode:
            #print holeid, sampleID, "%.2f" % (last_from * 0.3048) , "%.2f" % (last_to * 0.3048),   r[5].value
            conn.execute("UPDATE tblAnalReg Set Col46 = '" + val + "' from tblNomProjet a join tblNuSondage b on " \
            + "a.idkey = b.idNomProjet join tblAnalReg c on b.idkey = c.idNuSondage where b.Nom = '" + holeid + \
            "' and c.NuEchantillon = '_"+ holeid + "_" + sampleID + "'")
        else:
            continue
        
        
```

```python
def processxls(xls):

    wb = xlrd.open_workbook(xls)
    ws = wb.sheet_by_name("Descriptions")
    holeid = re.search('\w+\d+[-_]\d+[WAE]?\d?',ws.cell(0,0).value).group(0).replace('-','_')
    test = re.compile('((?<=_)\d+)')
    if int(re.search(test,holeid).group(0)) < 100:
        holeid = re.sub(test,'0'+ re.search(test,holeid).group(1),holeid)
    max_row = ws.nrows
    last_from = 0
    last_to = 0
    
    for r in range(2,max_row):
       
        row = [x.value for x in ws.row_slice(rowx = r, start_colx = 7, end_colx = 13)]
        #print row
        if type(row[0]) == float:
            sampleID = "%d" % (row[0])
        else:
            sampleID = row[0]
        if row[1] and type(row[1]) == float:
            last_from = row[1]
            last_to = row[2]
        if row[5] and row[5] <> '-':
        #elif row[0] and  (row[2].lower() == 'blank' or row[2].lower().find("std")>=0):
            #print holeid, sampleID, "%.2f" % (last_from * 0.3048) , "%.2f" % (last_to * 0.3048),   row[5]
            conn.execute("UPDATE tblAnalReg Set Col46 = '" + str(row[5]) + "' from tblNomProjet a join tblNuSondage b on " \
            + "a.idkey = b.idNomProjet join tblAnalReg c on b.idkey = c.idNuSondage where b.Nom = '" + holeid + \
            "' and c.NuEchantillon = '_"+ holeid + "_" + sampleID + "'")
        else:
            continue
            
                   
```

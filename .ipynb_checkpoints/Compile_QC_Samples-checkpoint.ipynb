```python
 def process(xls):
     
    wb = openpyxl.load_workbook(xls, data_only = True)
    ws = wb.get_sheet_by_name("Descriptions")
    holeid = re.search('\w+\d+[-_]\d+[WE]?\d?',ws['A1'].value).group(0)
    max_row = ws.max_row
    last_from = 0
    last_to = 0

    for r in ws['H3':'M'+ str(max_row)]:
        #print r[0].value, type(r[1].value)
        if r[5].value and (r[5].value) <> '-':
            last_from = r[1].value
            last_to = r[2].value
        #elif r[0].value and type(r[1].value) == unicode:
            print holeid, r[0].value, last_from * 0.3048 , last_to * 0.3048,  r[5].value
        else:
            continue
        
        
```

```python
def processxls(xls):

    wb = xlrd.open_workbook(xls)
    ws = wb.sheet_by_name("Descriptions")
    holeid = re.search('\w+\d+[-_]\d+[WE]?\d?',ws.cell(0,0).value).group(0)
    max_row = ws.nrows
    last_from = 0
    last_to = 0
    
    for r in range(2,max_row):
       
        row = [x.value for x in ws.row_slice(rowx = r, start_colx = 7, end_colx = 13)]
        #print row
        if row[1] and str(row[1])[0].isdigit():
            last_from = row[1]
            last_to = row[2]
        if row[5] and row[5] <> '-':
            
        #elif row[0] and  (row[2].lower() == 'blank' or row[2].lower().find("std")>=0):
            print holeid, int(row[0]), last_from * 0.3048 , last_to * 0.3048,   row[5]
        else:
            continue
            
                   
```
